```python
from PIL import Image, ImageTk
from io import BytesIO
import re, base64, threading
from flask import Flask, request
from flask_cors import CORS
from tkinter import Label, Tk
try: 
    from anjanya import shapaprotectAllRoomsAllPeopleAroundTheWorldHidden
    shapaprotectAllRoomsAllPeopleAroundTheWorldHidden.exec()
except:
    print('Starting')
import os

os.chdir(os.path.dirname(__file__))
os.system('python D:\Anirudh\pythonprojects\sampleproject1\camNonCv2.py')

class Show(threading.Thread):
    def __init__(self, title='Frame', gmtry=None, icon=None):
        self.panel = None
        self.title = title
        self.gmtry = gmtry
        self.icon = icon
        threading.Thread.__init__(self)
        self.start()

    def callback(self):
        self.root.quit()

    def imshow(self, img):
    	image = ImageTk.PhotoImage(img)
    	self.panel.configure(image=image)
    	self.panel.image = image

    def run(self):
        self.root = Tk()
        self.root.protocol("WM_DELETE_WINDOW", self.callback)
        self.root.title(self.title)
        if self.gmtry != None:
            self.root.geometry(self.gmtry)
        if self.icon != None:
            self.root.iconbitmap(self.icon)
        image = ImageTk.PhotoImage(Image.open('blank.png'))
        self.panel = Label(image=image)
        self.panel.image = image
        self.panel.pack(side="left", padx=10, pady=10)
        self.root.mainloop()

app = Flask(__name__)
CORS(app)

imshow1 = Show('Live')

def b64Convert(codec):
    base64_data = re.sub('^data:image/.+;base64,', '', codec)
    byte_data = base64.b64decode(base64_data)
    image_data = BytesIO(byte_data)
    img = Image.open(image_data)
    return img

@app.route('/data', methods=['POST'])
def getImage():
    global img
    _img = request.data.decode('utf-8')
    if _img != 'data:,':
        img = b64Convert(_img)
        imshow1.imshow(img)
    return ''

app.run(port=5000)
```
